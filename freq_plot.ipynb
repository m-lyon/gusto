```python
from gusto.experiments.unet.lib.dataset import UNetDataset

import matplotlib.pyplot as plt

import numpy as np

import torch
```

```python
dataset = UNetDataset(start=0.9, end=1.0, dims=(256, 256), strategy='wrap')
```

```python
example = dataset[0][0][0]
```

```python
plt.imshow(example[0, 0])
```

```python
# fft2 the example
example_fft = torch.fft.fft2(torch.Tensor(example))
example_fft = example_fft.numpy()
```

```python
plt.imshow(example_fft.imag[0, 0])
```

```python
plt.imshow(np.log(example_fft.real[0, 0]))
```
